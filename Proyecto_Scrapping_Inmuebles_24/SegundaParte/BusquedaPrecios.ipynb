```python
def convertir_a_string(df2):
    df = df2.copy()  # Crea una copia del DataFrame para evitar el warning
    df['NOMBRECOMERCIAL'] = df['NOMBRECOMERCIAL'].astype(str)
    df['FECHAREGISTRO'] = df['FECHAREGISTRO'].astype(str)
    df['PRESENTACIÓN'] = df['PRESENTACIÓN'].astype(str)
    df['MARCA'] = df['MARCA'].astype(str)
    df['PRECIO'] = df['PRECIO'].astype(float)
    df['LATITUD'] = df['LATITUD'].astype(float)
    df['LONGITUD'] = df['LONGITUD'].astype(float)
    return df

df_papa = convertir_a_string(df_papa)
df_lechuga = convertir_a_string(df_lechuga)
df_aguacate = convertir_a_string(df_aguacate)
df_jitomate = convertir_a_string(df_jitomate)
```

```python
def find_cheapest_store(datasets):
    store_prices = {}
    store_counts = {}
    
    for dataset in datasets:
        for index, entry in dataset.iterrows():
            store = entry['NOMBRECOMERCIAL']
            price = entry['PRECIO']
            if store in store_prices:
                store_prices[store] += price
                store_counts[store] += 1
            else:
                store_prices[store] = price
                store_counts[store] = 1
    
    num_datasets = len(datasets)
    valid_stores = {k: v for k, v in store_counts.items() if v == num_datasets}
    
    cheapest_store = min(valid_stores.keys(), key=lambda k: store_prices[k])
    
    return cheapest_store, store_prices[cheapest_store]
```

```python
datasets = [df_papa, df_lechuga, df_aguacate, df_jitomate]
cheapest_store, total_price = find_cheapest_store(datasets)
```

```python
def list_all_store_prices(datasets):
    store_prices = {}
    store_counts = {}
    
    for dataset in datasets:
        for index, entry in dataset.iterrows():
            store = entry['NOMBRECOMERCIAL']
            price = entry['PRECIO']
            if store in store_prices:
                store_prices[store] += price
                store_counts[store] += 1
            else:
                store_prices[store] = price
                store_counts[store] = 1
                
    return store_prices, store_counts

# Usar la función
datasets = [df_papa, df_lechuga, df_aguacate, df_jitomate]
all_store_prices, all_store_counts = list_all_store_prices(datasets)

# Imprimir los resultados
for store, total_price in all_store_prices.items():
    print(f"Tienda: {store}, Precio total: {total_price}, Número de productos: {all_store_counts[store]}")
```

```python
import folium

def list_all_store_info(datasets):
    store_info = {}
    
    for dataset in datasets:
        for index, entry in dataset.iterrows():
            store = entry['NOMBRECOMERCIAL']
            price = entry['PRECIO']
            lat = entry['LATITUD']
            lon = entry['LONGITUD']
            
            if store in store_info:
                store_info[store]['total_price'] += price
            else:
                store_info[store] = {'total_price': price, 'lat': lat, 'lon': lon}
    
    # Encuentra los límites para categorizar las tiendas.
    all_total_prices = [info['total_price'] for info in store_info.values()]
    low_price = min(all_total_prices)
    high_price = max(all_total_prices)
    mid_price = (low_price + high_price) / 2
    
    # Inicializar el mapa
    m = folium.Map(location=[20.0, -102.0], zoom_start=5)
    
    # Añadir las marcas al mapa
    for store, info in store_info.items():
        color = 'red'
        if info['total_price'] <= low_price + (mid_price - low_price) / 2:
            color = 'green'
        elif info['total_price'] <= mid_price:
            color = 'yellow'
        
        folium.Marker(
            [info['lat'], info['lon']],
            popup=f"{store}\nPrecio total: {info['total_price']}",
            icon=folium.Icon(color=color)
        ).add_to(m)
    
    return m

# Usar la función
datasets = [df_papa, df_lechuga, df_aguacate, df_jitomate]
m = list_all_store_info(datasets)

# Mostrar el mapa
m
```

```python
import folium

def list_all_store_info(store_datasets, rental_datasets):
    store_info = {}
    
    # Procesar los datasets de tiendas
    for dataset in store_datasets:
        for index, entry in dataset.iterrows():
            store = entry['NOMBRECOMERCIAL']
            price = entry['PRECIO']
            lat = entry['LATITUD']
            lon = entry['LONGITUD']
            
            if store in store_info:
                store_info[store]['total_price'] += price
            else:
                store_info[store] = {'total_price': price, 'lat': lat, 'lon': lon, 'type': 'store'}
    
    # Procesar los datasets de establecimientos para rentar
    for dataset in rental_datasets:
        for index, entry in dataset.iterrows():
            establishment = entry['ubicacion']
            rent_price = entry['precio']
            lat = entry['latitud']
            lon = entry['longitud']
            
            if establishment in store_info:
                store_info[establishment]['total_price'] += rent_price
            else:
                store_info[establishment] = {'total_price': rent_price, 'lat': lat, 'lon': lon, 'type': 'rental'}
    
    # Encuentra los límites para categorizar las tiendas y los establecimientos.
    all_total_prices = [info['total_price'] for info in store_info.values()]
    low_price = min(all_total_prices)
    high_price = max(all_total_prices)
    mid_price = (low_price + high_price) / 2
    
    # Inicializar el mapa
    m = folium.Map(location=[20.0, -102.0], zoom_start=5)
    
    # Añadir las marcas al mapa con íconos y colores diferentes
    for store, info in store_info.items():
        if info['type'] == 'store':
            price = info['total_price']
            if price < 150:
                color = 'green'
            elif 150 <= price <= 200:
                color = 'orange'
            else:
                color = 'red'
            
            icon = folium.Icon(color=color, icon='shopping-cart')
        elif info['type'] == 'rental':
            price = info['total_price']
            if price < 3000:
                color = 'green'
            elif 3000 <= price <= 8000:
                color = 'orange'
            else:
                color = 'red'
            
            icon = folium.Icon(color=color, icon='home')
        
        folium.Marker(
            [info['lat'], info['lon']],
            popup=f"{store}\nPrecio total: {info['total_price']}",
            icon=icon
        ).add_to(m)
    
    return m

# Usar la función
store_datasets = [df_papa, df_lechuga, df_aguacate, df_jitomate]
rental_datasets = [df2]
m = list_all_store_info(store_datasets, rental_datasets)

# Mostrar el mapa
m
```
